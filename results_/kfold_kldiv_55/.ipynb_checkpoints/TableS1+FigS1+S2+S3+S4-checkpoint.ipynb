```python
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import seaborn as sns
sns.set(context='talk', style='ticks',
        color_codes=True, rc={'legend.frameon': False})

%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.dpi'] = 300
#plt.subplots_adjust(wspace=0.25, hspace=0.3)

csv_names = sorted(glob('*.csv'))

fig, axs = plt.subplots(6,2, figsize=(15,25))
x_range = np.arange(0,160.01,20)
y_range = np.arange(0,5000.01,1000)

for n, csv_name in enumerate(csv_names):
    df = pd.read_csv(csv_name)    
    #df['AE'] = np.abs(df.Predicted - df.NIST)
    
    valid = df[df['Train/Valid/Test'] == 'Valid']
    test = df[df['Train/Valid/Test'] == 'Test']
    #print(csv_name, len(valid), valid.AE.mean())
    
    i,j = int(n/2), int(n%2)
    valid.NIST.hist(ax=axs[i][j], color='darkgreen', bins=x_range)
    
    axs[i][j].set_xticks(x_range)
    axs[i][j].set_yticks(y_range)
    axs[i][j].set_title('Fold number '+str(n+1))
    axs[i][j].set_xlabel('HoV (kJ/mol)')
    axs[i][j].set_ylabel('# of data points')
    
    #print(len(test))
    
test.NIST.hist(ax=axs[5][0], color='firebrick', bins=x_range)
axs[5][0].set_title('Test set')
axs[5][0].set_xticks(x_range)
axs[5][0].set_yticks(y_range)
axs[5][0].set_xlabel('HoV (kJ/mol)')
axs[5][0].set_ylabel('# of data points')

df.NIST.hist(ax=axs[5][1], color = 'black', bins=x_range)
axs[5][1].set_title('Whole database')
axs[5][1].set_xticks(x_range)
#axs[5][1].set_yticks(y_range)
axs[5][1].set_xlabel('HoV (kJ/mol)')
axs[5][1].set_ylabel('# of data points')


fig.tight_layout()
```

```python
unc_mean = df[df.DB_unc > 0.01].DB_unc.mean()
df[df.DB_unc > 0.01].DB_unc.hist(bins=np.arange(0,16.01,0.5))
plt.axvline(unc_mean, color='k', linestyle='dashed', linewidth=2)
plt.text(unc_mean*1.1, 25500, 'Mean: {:.2f}'.format(unc_mean), fontsize=16)
```

```python
matplotlib.rcParams.update({'errorbar.capsize': 3})

labels = ['(1,1)', '(1,2)','(1,3)', '(1,4)', 
          '(2,1)', '(2,2)','(2,3)', '(2,4)', 
          '(3,1)', '(3,2)','(3,3)', '(3,4)']
          #'(4,1)', '(4,2)','(4,3)', '(4,4)',
          #'(5,4)', '(5,5)', '(5,6)']
MSE_loss = [5.24,5.15,5.08,5.0,
            5.1,4.99,4.91,4.88,
           5.07,4.87,4.84,4.88]

           #4.96,4.88,4.91,4.79,
           #4.77,4.76,4.72]
KL_div_loss = [5.65,5.35,5.19,5.38,
            5.31,4.98,5.24,5.0,
           5.1,4.8,5.44,4.73]
           #5.19,4.74,4.71,4.9,
           #4.78,4.49,4.63]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(11,4))
rects1 = ax.bar(x - width/2, MSE_loss, width, label='Loss - MSE') 
rects2 = ax.bar(x + width/2, KL_div_loss, width, label='Loss - KL_Div')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('(# of layers, # of attention heads)')
ax.set_ylabel('MAE (kJ/mol)')
#ax.set_title('')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim([4.0,6.0])
ax.set_yticks(np.arange(4.0,6.01,0.4))
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{0:.2f}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
