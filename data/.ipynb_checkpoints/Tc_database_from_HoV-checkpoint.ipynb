```python
import rdkit.Chem
import numpy as np
from rdkit.Chem.GraphDescriptors import Chi1,Ipc
from rdkit.Chem.EState.EState import EStateIndices
from rdkit.Chem.rdMolDescriptors import CalcNumRings, CalcNumHBD

def glob_features(smi):
    mol = rdkit.Chem.MolFromSmiles(smi)
    return np.array([ Chi1(mol), Ipc(mol), sum(EStateIndices(mol)), CalcNumRings(mol), CalcNumHBD(mol)   ])
    #return np.array([ CalcNumRings(mol), CalcNumHBD(mol)   ])  
```

```python
import pandas as pd
data = pd.read_csv('Tc.csv')
data['glob_feat'] = data['smiles'].apply(lambda x: glob_features(x))
```

```python
import tensorflow as tf
train_data = data.sample(frac=.8, random_state=1)
#gf = tf.constant(list(data.glob_feat))
gf = tf.constant(list(train_data.glob_feat))
gf.shape
```

```python
from tensorflow.keras.layers.experimental import preprocessing

layer = preprocessing.Normalization()
layer.adapt(gf)
normalized_data = layer(gf)

print(normalized_data.numpy())

print("Features mean: %.2f" % (normalized_data.numpy().mean()))
print("Features std: %.2f" % (normalized_data.numpy().std()))
```

```python
#tmp = np.array(list(list(data.glob_feat))).T
tmp = np.array(list(list(train_data.glob_feat))).T

for i in range(len(tmp)):
    mean = tmp[i].mean()
    std = tmp[i].std()
    tmp[i] = (tmp[i] - mean) / std

print(tmp.T)
```
