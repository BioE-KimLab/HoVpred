```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(context='talk', style='ticks',
        color_codes=True, rc={'legend.frameon': False})

%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.dpi'] = 300
plt.rcParams["font.family"] = 'Arial'
plt.rcParams.update({'font.size': 24})
import numpy as np


result = pd.read_csv('FP_best/results.csv')

train_x = result[result['Train/Valid/Test'] == 'Train']['Predicted']
train_y = result[result['Train/Valid/Test'] == 'Train']['DB_value']
valid_x = result[result['Train/Valid/Test'] == 'Valid']['Predicted']
valid_y = result[result['Train/Valid/Test'] == 'Valid']['DB_value']
test_x = result[result['Train/Valid/Test'] == 'Test']['Predicted']
test_y = result[result['Train/Valid/Test'] == 'Test']['DB_value']

fig = plt.figure(figsize=(5.5,5.5))
ax = fig.add_subplot(111)
ax.axis('equal')
#ax.set_xticks(np.arange(0.0,240.01,40.0))
#ax.set_yticks(np.arange(0.0,240.01,40.0))
ax.set_xlim([200,600])
ax.set_ylim([200,600])
ax.set_xlabel('FP (Predicted, K)')
ax.set_ylabel('FP (Database, K)')
plt.text(215,575,'[MAE, K]', fontdict={"size":16})

ax.plot(train_x, train_y, '.', ms=10, label='Training [6.4]', alpha=0.5, color='navy')
ax.plot(valid_x, valid_y, '.',  ms=10, label='Validation [7.1]', alpha=1.0, color='darkgreen')
ax.plot(test_x, test_y, '.',  ms=10, label='Test [6.5]', alpha=0.5, color='firebrick')
lgnd = ax.legend(loc='center left', bbox_to_anchor=(-0.05, .8))

for lh in lgnd.legendHandles: 
    lh._legmarker.set_alpha(1)
    lh._legmarker.set_markersize(15)

ax.plot([200,600], [200, 600], '--', color='.1')
fig.savefig('FP_parity_220702.tiff', format='tiff', dpi=300)
```

```python
result = pd.read_csv('FP_wholeDB_transfer_learning/fp_results.csv')

train_x = result[result['Train/Valid/Test'] == 'Train']['Predicted']
train_y = result[result['Train/Valid/Test'] == 'Train']['FP']
valid_x = result[result['Train/Valid/Test'] == 'Valid']['Predicted']
valid_y = result[result['Train/Valid/Test'] == 'Valid']['FP']
test_x = result[result['Train/Valid/Test'] == 'Test']['Predicted']
test_y = result[result['Train/Valid/Test'] == 'Test']['FP']

fig = plt.figure(figsize=(5.5,5.5))
ax = fig.add_subplot(111)
ax.axis('equal')
ax.set_xticks(np.arange(100.0,750.01,130.0))
ax.set_yticks(np.arange(100.0,750.01,130.0))
ax.set_xlim([100,750])
ax.set_ylim([100,750])
ax.set_xlabel('FP (Predicted, K)')
ax.set_ylabel('FP (Database, K)')
plt.text(130,710,'[MAE, K]', fontdict={"size":16})

ax.plot(train_x, train_y, '.', ms=10, label='Training [16.3]', alpha=0.5, color='navy')
ax.plot(valid_x, valid_y, '.',  ms=10, label='Validation [15.6]', alpha=1.0, color='darkgreen')
ax.plot(test_x, test_y, '.',  ms=10, label='Test [14.7]', alpha=0.5, color='firebrick')
lgnd = ax.legend(loc='center left', bbox_to_anchor=(-0.07, .8))
    
for lh in lgnd.legendHandles: 
    lh._legmarker.set_alpha(1)
    lh._legmarker.set_markersize(15)
    
ax.plot([100,750], [100, 750], '--', color='.1')
```
